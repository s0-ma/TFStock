```python
# 人間でも分類できるダミーデータ
list_df = []
list_df.append(pd.read_csv('../data/quadratic_0.5/-2_001.csv', index_col="Date"))
list_df.append(pd.read_csv('../data/quadratic_0.5/-1_001.csv', index_col="Date"))
list_df.append(pd.read_csv('../data/quadratic_0.5/0_001.csv', index_col="Date"))
list_df.append(pd.read_csv('../data/quadratic_0.5/1_001.csv', index_col="Date"))
list_df.append(pd.read_csv('../data/quadratic_0.5/2_001.csv', index_col="Date"))

fig, (axU, axB) = plt.subplots(nrows=2, sharex=True, sharey=True)
for index, df in enumerate(list_df):
    df['Adj Close'].plot(ax=axU)
    y = -2.5 + index
    axU.axhline(y=y, linestyle='dashed', color='0.5')
    
for index, df in enumerate(list_df):
    df['Adj Close'][:30].plot(ax=axB)
    y = -2.5 + index
    axB.axhline(y=y, linestyle='dashed', color='0.5')

plt.ylim((-2.5,2.5))
plt.xlim((0, 60))
plt.show()
```

```python
# 人間でも分類できるダミーデータ
import random

list_df = []
list_df.append([0,  pd.read_csv('../data/quadratic_0.5/0_041.csv',  index_col="Date")])
list_df.append([1,  pd.read_csv('../data/quadratic_0.5/1_031.csv',  index_col="Date")])
list_df.append([-2, pd.read_csv('../data/quadratic_0.5/-2_192.csv', index_col="Date")])
random.shuffle(list_df)

def plot_df(df):
    df['Adj Close'][:30].plot()
    for i in range(5):
        y = -2.5 + i
        plt.axhline(y=y, linestyle='dashed', color='0.5')
    plt.ylim((-2.5,2.5))
    plt.xlim((0, 60))
    plt.show()

plot_df(list_df[0][1])
plot_df(list_df[1][1])
plot_df(list_df[2][1])
```
